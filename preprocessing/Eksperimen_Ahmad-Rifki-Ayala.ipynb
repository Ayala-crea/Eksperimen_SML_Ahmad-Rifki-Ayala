```python
df_model = df.copy()

drop_cols = [
    "ID",
    "submission_of_application",  # jika tidak diolah jadi fitur waktu
]

df_model = df_model.drop(columns=[c for c in drop_cols if c in df_model.columns], errors="ignore")

print("Shape setelah drop:", df_model.shape)
```

```python
df_check = df_model.copy()

# Cek jumlah missing value per kolom
missing_summary = df_check.isnull().sum()

print("Jumlah Missing Value per Kolom:")
print(missing_summary)

# (Opsional) tampilkan hanya kolom yang ada missing value
print("\nKolom yang memiliki Missing Value:")
print(missing_summary[missing_summary > 0])
```

```python
df_imputed = df_model.copy()

# --- 1) (Opsional) ringkas missing sebelum ---
missing_before = df_imputed.isnull().sum()
print("Missing sebelum (kolom > 0):")
print(missing_before[missing_before > 0].sort_values(ascending=False))

# --- 2) Tentukan kolom numeric vs categorical ---
num_cols = df_imputed.select_dtypes(include=[np.number]).columns
cat_cols = df_imputed.select_dtypes(exclude=[np.number]).columns

# --- 3) Imputasi numeric -> median ---
for c in num_cols:
    if df_imputed[c].isnull().any():
        med = df_imputed[c].median()
        # kalau satu kolom semuanya NaN, median jadi NaN -> fallback 0
        if pd.isna(med):
            med = 0
        df_imputed[c] = df_imputed[c].fillna(med)

# --- 4) Imputasi categorical -> mode (most frequent) ---
for c in cat_cols:
    if df_imputed[c].isnull().any():
        modes = df_imputed[c].mode(dropna=True)
        fill_val = modes.iloc[0] if len(modes) > 0 else "Unknown"
        df_imputed[c] = df_imputed[c].fillna(fill_val)

# --- 5) Validasi setelah imputasi ---
missing_after = df_imputed.isnull().sum()
print("\nMissing sesudah (kolom > 0):")
print(missing_after[missing_after > 0])
```

```python
duplicate_count = df_imputed.duplicated().sum()

print("Jumlah data duplikat:", duplicate_count)
```

```python
df_scaled = df_imputed.copy()

target_col = "Status"

# Pisahkan fitur & target
X = df_scaled.drop(columns=[target_col])
y = df_scaled[target_col]

# Ambil kolom numerik saja
num_cols = X.select_dtypes(include=[np.number]).columns

# Inisialisasi scaler
scaler = StandardScaler()

# Fit & transform hanya kolom numerik
X[num_cols] = scaler.fit_transform(X[num_cols])

# Gabungkan kembali
df_scaled = pd.concat([X, y], axis=1)

print("Scaling selesai.")
print("Shape:", df_scaled.shape)
df_scaled.head()
```

```python
df_outlier_check = df_scaled.copy()

# Ambil kolom numerik saja (tanpa target)
num_cols = df_outlier_check.drop(columns=["Status"]).select_dtypes(include=[np.number]).columns

outlier_summary = {}

for col in num_cols:
    Q1 = df_outlier_check[col].quantile(0.25)
    Q3 = df_outlier_check[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df_outlier_check[(df_outlier_check[col] < lower_bound) |
                                 (df_outlier_check[col] > upper_bound)]

    outlier_summary[col] = len(outliers)

# Tampilkan kolom yang punya outlier
outlier_df = pd.Series(outlier_summary)
print("Jumlah Outlier per Kolom:")
print(outlier_df[outlier_df > 0].sort_values(ascending=False))
```

```python
df_capped = df_scaled.copy()

target_col = "Status"
num_cols = df_capped.drop(columns=[target_col]).select_dtypes(include=[np.number]).columns

LOW_Q = 0.01   # 1%
HIGH_Q = 0.99  # 99%

cap_info = []

for col in num_cols:
    lower = df_capped[col].quantile(LOW_Q)
    upper = df_capped[col].quantile(HIGH_Q)

    before_outliers = ((df_capped[col] < lower) | (df_capped[col] > upper)).sum()

    df_capped[col] = df_capped[col].clip(lower=lower, upper=upper)

    after_outliers = ((df_capped[col] < lower) | (df_capped[col] > upper)).sum()

    cap_info.append([col, lower, upper, int(before_outliers), int(after_outliers)])

cap_df = pd.DataFrame(cap_info, columns=["col", "lower", "upper", "outliers_before", "outliers_after"])
print(cap_df.sort_values("outliers_before", ascending=False).head(20))
```

```python
df_ready = df_capped.copy()
target_col = "Status"

# Pisahkan X dan y
X = df_ready.drop(columns=[target_col])
y = df_ready[target_col].astype(int)

# Deteksi kolom categorical (object)
cat_cols = X.select_dtypes(include=["object"]).columns.tolist()

print("Jumlah kolom categorical:", len(cat_cols))
print("Contoh categorical cols:", cat_cols[:10])

# One-Hot Encoding
X_encoded = pd.get_dummies(
    X,
    columns=cat_cols,
    drop_first=True,     # opsional: mengurangi dummy trap
    dummy_na=False       # karena missing sudah di-impute
)

# Gabungkan lagi jadi df final untuk modeling
df_final = pd.concat([X_encoded, y], axis=1)

print("Shape sebelum encoding:", X.shape)
print("Shape sesudah encoding:", X_encoded.shape)
print("Final shape:", df_final.shape)
```

```python
df_final.to_csv("dataset_preprocessed_final.csv", index=False)
print("Dataset berhasil disimpan sebagai dataset_preprocessed_final.csv")
```
